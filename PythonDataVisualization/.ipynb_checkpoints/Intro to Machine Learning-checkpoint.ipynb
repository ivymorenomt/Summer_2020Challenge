# Machine Learning 101

- the process of deploying computer power to identify trends and patterns in data in order to make predictions about how that data will behave in the future

## Machine Learning use cases
- sales forecasting
- customer segment analysis
- insurance claim fraud detection
- hedge fund classification
- spam detection

## Application
- Adaptive websites
- speed and handwriting recognition
- etc..

### The models can be used to
- identify relationships between variables(regression)
- assign new data to the correct group(classification)
- Find the distribution of inputs in a certain space(density estimation)
- simplify inputs by mappin them into a lower dimensional space(dimensionality reduction)
- Group similar data points together, without knowing the groups beforehand(clustering)

## Machine Learning
- it is a field of computer science where algorithms learn from data. The field is a combination of
    * computational statistics
    * Pattern Recognition
    * Mathematical optimization
    * computational learning theory in AI
- it involves data mining and data exploring analysis

### Types of ML Algorithms by function
- regression - it enables us to studt labeled data and quantify linear relationships between at least one predictor variable (X) and one response (Y)
    - linear
    - logistic
- clustering - takes unlabeled, categorical data and organizes it into clusters based on similarity in attribute values of the data points
    - k-means clustering
    - hierarchical methods
    - DBSCAN
- dimension reduction - takes unlabeled, continuous, high dimensional data and compresses it into a smaller synthetic representation
    - explanatory factor analysis
    - principal component analysis(PCA)
- association rules
    - association rules with Apriori
- deep learning
    - Perceptron
- instance-based - classification learns from labeled data to predict categories for new data points; the class labels are discrete, unordered values
    - K-nearest neighbor Algorithm(KNN)
- decision trees
    - CART
- bayesian
    - Naive Bayes
- ensemble
    - Random Forest
- regularization

### ML Lingo
Feature: is synonymous with terms variable, column, attribute and field
Instance: synonymous with the terms row, observation, data point, value, and case(dataset)
Target: a target variable is synonymous with the terms predictant and dependent variable
Data: the term is used synonymously to refer to a predictor or set of predictor variables - this is what we fee to the machine

### ML Learning Test and Training Sets

Use random sampling to generate random samples and then break the data into test and training sets
* 4/5 of your dataset -> Training set(use it to train your model)
* 1/5 of your dataset -> Test Set(for testing to see how well the model performs)

### Types of ML Methods
- Supervised - Making predictions straight from labeled data
    - example: an apple, orange and banana are the datasets and name it with appropriate label
    - then someone comes in with unknown fruit and asked what is this
    - you may guess it is an orange, due to skin type, color, texture..
- Unsupervised - Making predictions straight from unlabeled data
    - example: they give you 7 fruits unknown and give you 8th.
    - try to find the subgroups you were originally given - which is citrus, rose family, and berries. 
    - similarities of the skin
    - then go back to the unknown fruit. which group would it fall into
- Semi-Supervised - Uses both labeled and unlabeled data to make a set of predictions


### Benefits of understanding ML Groupings
- total ML Algorithms > 100
- total ML Use cases > 100

It's important to understand ways to group them in order to decrease time-to-value by hastening the model selection process

### Popular ways to group ML Algorithms
- By Learning
- By Fnction
- By Use case
### Grouping by Learning style
- if your data is continuous and labeled: Supervised Learning > Regression
- if your data is categorical and labeled: Supervised Learning > Classification
- if your data is continuous and unlabeled: Unsupervised Learning > Dimensionality reduction
- if your data is categorical and unlabeled: Unsupervised Learning > Clustering
