# Three Different Data Visualization Types
- data storytelling - for presentations to organizational decision makers
- data showcasing - for presentations to analysts, scientist, mathematicians, and engineers
- data art - for presentations to activists or to the general public

### Data Storytelling
Make it easy for the audience to get the point
Your data visualization: Clutter free, highly organized
Intended audience - non analyst and non technical business managers

Product types
- static images
- simple interactive dashboards

Example = global warming. what is making it warm? it should be understandable

### Data Showcasing
Showcase lots of data so your audience members can think for themselves
Visualization: highly contextual, open ended(allowing them to think for themselves)
Intended audience - quants, analysts, engineers, scientists
Product types
- Static images
- interactive dashboards

Example: data science for dummies - look for trends, patterns and correlations. Good for exploratory data analysis on a visual basis

### Data Art
- use to make a statement
- activist to take an action
Visualization: Attention getting, creative, controversial
Intnded audience: idealists, social activists, artist
Product types:
- static images
Example: from periscopic art - it displays gun killings in the US
https://www.periscopic.com - instead of taking lives lost, they use lives stolen

### Defining axes, ticks and grids

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
ax.plot(x,y)
```

```python
#Set ticks/add limits to the ticks
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5])

ax.plot(x,y)
```

```python
#add grid
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.grid()
ax.plot(x,y)
```

### Create Statistical Data Graphics
Statistical plots allow viewers to
- identify outliers
- visualize distributions
- deduce variable types
- discover relationships and core relations between variables in a dataset

#### Histograms
- it shows a variable's distribution as a set of adjacent rectangles on a data chart. It represent counts of data within a numerical range of values

#### Scatterplots
- it is useful when you want to explore interrelations or dependencies between two different variables. These data graphics are ideal for visually spotting outliers and trends in data.

#### Boxplots
- it is useful for seeing a variable's spread and for detecting outliers

```python
import seaborn as sb
sb.set_style('whitegrid')
```

```python
cars.index = cars.car_names

mpg = cars['mpg']
mpg.plot(kind='hist')
```

```python
# Create a histogram using Seaborn
sb.distplot(mpg) 
```

```python
cars.plot(kind='scatter', x='hp', y='mpg', c=['darkgray'], s=150)
```

```python
#Scatterplot with seaborn
sb.regplot(x='hp', y='mpg', data=cars, scatter=True)
```

```python
#Generate a scatter plot matrix
# running this code is taking a while
sb.pairplot(cars)
```

```python
# A performance lag is noticed. It will take a while to show the graphs using pairplot
cars_subset = cars[['mpg', 'disp', 'hp', 'wt']]
sb.pairplot(cars_subset)
plt.show()
```

```python
# Boxplots
cars.boxplot(column = 'mpg', by='am')
cars.boxplot(column = 'wt', by='am')

# if the car is a automatic transmission, mpg is better
# If the car is not a automatic transmission, then it tends to be having more weight 
```

```python
#Create boxplot in Seaborn
sb.boxplot(x='am', y='mpg', data=cars, palette='hls')
```
